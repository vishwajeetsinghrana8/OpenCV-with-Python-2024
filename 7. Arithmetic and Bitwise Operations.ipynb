# **Arithmetic and Bitwise Operations**

Image arithmetic and bitwise operations are fundamental techniques in image processing used to perform mathematical operations and logical operations on pixel values in images. These operations are often used for tasks such as image blending, masking, thresholding, and bitwise logical operations. Let's delve into each concept:

**1. Image Arithmetic:**

Image arithmetic involves performing arithmetic operations such as addition, subtraction, multiplication, and division on corresponding pixel values of two images. These operations are typically performed on images of the same size and shape.

- **Addition and Subtraction:** When two images are added or subtracted pixel-wise, the corresponding pixel values are added or subtracted from each other. This can be useful for tasks like image blending and contrast adjustment.

- **Multiplication and Division:** Multiplying or dividing corresponding pixel values of two images results in a new image where each pixel value is the product or quotient of the corresponding pixel values in the original images. These operations can be useful for tasks such as intensity transformations and normalization.

**2. Bitwise Operations:**

Bitwise operations involve performing logical operations on the binary representations of pixel values in images. These operations are typically used for tasks such as image masking, thresholding, and creating binary images.

- **AND Operation:** In the bitwise AND operation, the corresponding bits of the binary representations of pixel values in two images are compared, and the result is a new image where each pixel value is the bitwise AND of the corresponding pixel values in the original images. This operation is often used for masking and image segmentation.

- **OR Operation:** The bitwise OR operation compares the corresponding bits of pixel values in two images and produces a new image where each pixel value is the bitwise OR of the corresponding pixel values in the original images. This operation is useful for combining masks and creating composite images.

- **NOT Operation:** The bitwise NOT operation inverts the bits of pixel values in an image, resulting in a new image where each pixel value is the bitwise complement of the corresponding pixel value in the original image. This operation is useful for creating binary masks and inverting image regions.

- **XOR Operation:** The bitwise XOR (exclusive OR) operation compares the corresponding bits of pixel values in two images and produces a new image where each pixel value is the bitwise XOR of the corresponding pixel values in the original images. This operation can be useful for tasks such as feature extraction and image differencing.

These image arithmetic and bitwise operations are essential tools in the image processing toolbox, providing powerful mechanisms for manipulating and analyzing image data. They are widely used in various applications, including computer vision, pattern recognition, and image analysis.

```python
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 7):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

## **Arithmetic Operations**

These are simple operations that allow us to directly add or subract to the color intensity.

Calculates the per-element operation of two arrays. The overall effect is increasing or decreasing brightness.

```python
# Adding comma zero in cv2.imread loads our image in as a grayscaled image
image = cv2.imread('image/lena.png', 0)
imshow("Grayscaled Image",image)
print(image)

# Create a matrix of ones, then multiply it by a scaler of 100 
# This gives a matrix with same dimesions of our image with all values being 100
M = np.ones(image.shape, dtype = "uint8") * 100 
print(image.shape)
```

```python
print(M)
print(M.shape)
```

#### **Increasing Brightness**

```python
# We use this to add this matrix M, to our image
# Notice the increase in brightness
added = cv2.add(image, M)
imshow("Increasing Brightness", added)
# Now if we just added it, look what happens
added2 = image + M 
imshow("Simple Numpy Adding Results in Clipping", added2)
```

#### **Decreasing Brightness**

```python
# Likewise we can also subtract
# Notice the decrease in brightness
subtracted = cv2.subtract(image, M)
imshow("Subtracted", subtracted)

subtracted = image - M 
imshow("Subtracted 2", subtracted)
```

## **Bitwise Operations and Masking**

To demonstrate these operations let's create some simple images

```python
# If you're wondering why only two dimensions, well this is a grayscale image, 
# Making a square
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
imshow("square", square)

# Making a ellipse
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
imshow("ellipse", ellipse)
```

### **Experimenting with some bitwise operations such as AND, OR, XOR and NOT**

```python
# Shows only where they intersect
And = cv2.bitwise_and(square, ellipse)
imshow("AND", And)

# Shows where either square or ellipse is 
bitwiseOr = cv2.bitwise_or(square, ellipse)
imshow("bitwiseOr", bitwiseOr)

# Shows where either exist by itself
bitwiseXor = cv2.bitwise_xor(square, ellipse)
imshow("bitwiseXor", bitwiseXor)

# Shows everything that isn't part of the square
bitwiseNot_sq = cv2.bitwise_not(square)
imshow("bitwiseNot_sq", bitwiseNot_sq)

# Notice the last operation inverts the image totally
```
