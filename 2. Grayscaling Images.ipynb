# **Grayscaling Images**

Converting an RGB (Red, Green, Blue) image to grayscale is a common preprocessing step in computer vision and image processing tasks. There are several reasons why this conversion is performed:

1. **Dimensionality Reduction**: Grayscale images have only one channel (intensity), while RGB images have three channels (red, green, blue). Converting RGB images to grayscale reduces the dimensionality of the image, making it easier and more efficient to process.

2. **Simplicity**: Grayscale images are simpler to work with computationally compared to RGB images. They require less memory and processing power, which can be beneficial for tasks with limited computational resources.

3. **Focus on Image Intensity**: In many cases, the color information in an image is not necessary for the analysis or task at hand. By converting to grayscale, the focus shifts to the intensity variations in the image, which often contain important visual information such as edges, textures, and shapes.

4. **Uniformity**: Grayscale images represent all colors using shades of gray, where darker shades correspond to lower intensity values and lighter shades correspond to higher intensity values. This uniform representation can simplify certain image processing tasks, such as edge detection and segmentation.

5. **Compatibility**: Some image processing algorithms and techniques are designed to work specifically with grayscale images. Converting RGB images to grayscale allows you to leverage these algorithms and techniques seamlessly.

6. **Reducing Redundancy**: In some cases, the color information in an RGB image may contain redundant or correlated data. Converting to grayscale can help reduce redundancy and improve the efficiency of subsequent processing steps.

Overall, converting RGB images to grayscale is a preprocessing step that can enhance the efficiency, simplicity, and effectiveness of various computer vision and image processing tasks, depending on the specific requirements of the application.

In this lesson we'll learn to:
1. Convert a color image to grayscale
2. See the change in dimensions between grayscale and color images



```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Define our imshow function 
def imshow(title = "Image", image = None, size = 8):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Load our input image
image = cv2.imread('./image/lena.png')

imshow("Lena", image)
```

```python
# We use cvtColor, to convert to grayscale
# It takes 2 arguments, the first being the input image
# The second being the color space conversion code 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Converted to Grayscale", gray_image)
```
