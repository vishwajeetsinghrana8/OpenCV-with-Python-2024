# **Thresholding, Binarization & Adaptive Thresholding**

Image thresholding is a fundamental technique in computer vision used to segment images into regions based on pixel intensity values. The goal of thresholding is to separate objects or features of interest from the background by binarizing the image, i.e., converting it into a binary image where each pixel is assigned either a foreground (object) or background (non-object) value.

Thresholding involves comparing the intensity of each pixel in the original image to a predefined threshold value. If the intensity of a pixel exceeds the threshold, it is classified as part of the foreground; otherwise, it is classified as part of the background.

There are several types of thresholding methods, each with its own characteristics and applications:

1. **Global Thresholding:** In global thresholding, a single threshold value is applied to the entire image to separate the foreground from the background. This method assumes that the intensity distribution of the entire image is uniform and does not take into account local variations in intensity. Global thresholding is simple and computationally efficient but may not be effective for images with uneven lighting or varying intensity levels.

2. **Adaptive Thresholding:** Adaptive thresholding overcomes the limitations of global thresholding by dynamically adjusting the threshold value based on local image properties. Instead of applying a single threshold to the entire image, adaptive thresholding computes a separate threshold for each pixel neighborhood, taking into account variations in intensity within the image. Adaptive thresholding is particularly effective for images with non-uniform illumination or varying contrast levels.

3. **Otsu's Thresholding:** Otsu's thresholding is an automatic thresholding method that calculates the optimal threshold value to minimize the intra-class variance of pixel intensities in the foreground and background regions. Otsu's method maximizes the separability between the foreground and background classes, making it suitable for images with bimodal intensity distributions.

4. **Histogram-based Thresholding:** Histogram-based thresholding methods analyze the histogram of pixel intensities in the image to determine an appropriate threshold value. These methods can be used to automatically select thresholds based on specific criteria, such as minimizing entropy, maximizing between-class variance, or optimizing image segmentation performance.

5. **Color Thresholding:** In addition to intensity-based thresholding, color thresholding can be applied to multi-channel color images by thresholding each color channel separately or by converting the image to a different color space (e.g., HSV or LAB) and thresholding based on color components such as hue, saturation, or chrominance.

Image thresholding is a versatile and powerful technique in computer vision, commonly used for tasks such as object detection, image segmentation, edge detection, and feature extraction. By partitioning an image into binary regions, thresholding provides a foundation for further analysis and processing of visual information.

```python
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 7):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Load our image as greyscale 
image = cv2.imread('image/lena.png',0)
imshow("Original", image)

# Values below 127 goes to 0 or black, everything above goes to 255 (white)
ret,thresh1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
imshow('1 Threshold Binary @ 127', thresh1)
print(ret)

# Values below 127 go to 255 and values above 127 go to 0 (reverse of above)
ret,thresh2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
imshow('2 Threshold Binary Inverse @ 127', thresh2)

# Values above 127 are truncated (held) at 127 (the 255 argument is unused)
ret,thresh3 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
imshow('3 THRESH TRUNC @ 127', thresh3)

# Values below 127 go to 0, above 127 are unchanged  
ret,thresh4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
imshow('4 THRESH TOZERO @ 127', thresh4)

# Reverse of the above, below 127 is unchanged, above 127 goes to 0
ret,thresh5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)
imshow('5 THRESH TOZERO INV @ 127', thresh5)
```

#### **cv2.adaptiveThreshold Parameters**

``**cv2.adaptiveThreshold**(src, maxValue, adaptiveMethod, thresholdType, blockSize, C[, dst]) → dst``

- **src** – Source 8-bit single-channel image.
- **dst** – Destination image of the same size and the same type as src .
- **maxValue** – Non-zero value assigned to the pixels for which the condition is satisfied. See the details below.
- **adaptiveMethod** – Adaptive thresholding algorithm to use, ADAPTIVE_THRESH_MEAN_C or ADAPTIVE_THRESH_GAUSSIAN_C . See the details below.
- **thresholdType** – Thresholding type that must be either THRESH_BINARY or THRESH_BINARY_INV .
- **blockSize** – Size of a pixel neighborhood that is used to calculate a threshold value for the pixel: 3, 5, 7, and so on.
- **C** – Constant subtracted from the mean or weighted mean. Normally, it is positive but may be zero or negative as well.


```python
image = cv2.imread('image/lena.png',0)
imshow("Original", image)

# Values below 127 goes to 0 (black, everything above goes to 255 (white)
ret,thresh1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
imshow('Threshold Binary', thresh1)

# It's good practice to blur images as it removes noise
image = cv2.GaussianBlur(image, (3, 3), 0)

# Using adaptiveThreshold
thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 5) 
imshow("Adaptive Mean Thresholding", thresh) 

_, th2 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow("Otsu's Thresholding", th2) 

# Otsu's thresholding after Gaussian filtering
blur = cv2.GaussianBlur(image, (5,5), 0)
_, th3 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow("Guassian Otsu's Thresholding", th3) 
```

### **SkImage Threshold Local**


``threshold_local(image, block_size, offset=10)``

The threshold_local function, calculates thresholds in regions with a characteristic size ``block_size`` surrounding each pixel (i.e. local neighborhoods). Each threshold value is the weighted mean of the local neighborhood minus an ``offset`` value


https://scikit-image.org/docs/stable/auto_examples/applications/plot_thresholding.html

```python
from skimage.filters import threshold_local

image = cv2.imread('image/lena.png')

# We get the Value component from the HSV color space 
# then we apply adaptive thresholdingto 
V = cv2.split(cv2.cvtColor(image, cv2.COLOR_BGR2HSV))[2]
T = threshold_local(V, 25, offset=15, method="gaussian")

# Apply the threshold operation 
thresh = (V > T).astype("uint8") * 255
imshow("threshold_local", thresh)
```
